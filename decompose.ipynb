```r
dataset<-read.csv(file.choose())
```

```r
temp<-dataset[c(2,5)]
```

```r
dataset<-ts(temp$Value, start=c(1948,1), freq=12)
```

```r
head(dataset)
```

```r
plot(dataset, type="l")
```

```r
datasetdecompose<-decompose(dataset)
```

```r
plot(datasetdecompose)
```

```r
datasetseasadj<- dataset - datasetdecompose$seasonal
```

```r
plot(datasetseasadj)
```

```r
datasetdetrend<- dataset - datasetdecompose$trend
```

```r
plot(datasetdetrend)
```

```r
finaldataset<- dataset - datasetdecompose$seasonal - datasetdecompose$trend
```

```r
plot(finaldataset)
```

```r
random<- dataset - datasetdecompose$random
```

```r
plot(random)
```
